```python
customers_df = df[[
    'customer_id',
    'customer_city',
    'customer_state',
    'customer_tier',
    'customer_spending_tier',
    'customer_age_group',
    'is_prime_member'
]].drop_duplicates(subset=['customer_id'])

customers_df.to_sql('customers', engine, if_exists='append', index=False)
```

```python
products_df = df[[
    'product_id',
    'product_name',
    'category',
    'subcategory',
    'brand',
    'product_weight_kg',
    'product_rating',
    'is_prime_eligible'
]].drop_duplicates(subset=['product_id'])

products_df.to_sql('products', engine, if_exists='append', index=False)
```

```python
time_df = df[[
    'order_date',
    'order_month',
    'order_year',
    'order_quarter'
]].drop_duplicates(subset=['order_date'])

time_df.to_sql('time_dimension', engine, if_exists='append', index=False)
```

```python
transactions_df = df[[
    'transaction_id',
    'customer_id',
    'product_id',
    'order_date',
    'quantity',
    'original_price_inr',
    'discount_percent',
    'discounted_price_inr',
    'subtotal_inr',
    'delivery_charges',
    'final_amount_inr',
    'payment_method',
    'delivery_days',
    'delivery_type',
    'is_festival_sale',
    'festival_name',
    'customer_rating',
    'return_status'
]]

transactions_df.to_sql('transactions', engine, if_exists='append', index=False)
```
