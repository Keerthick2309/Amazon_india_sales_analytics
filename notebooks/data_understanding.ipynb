- Merged all the 11 CSV to one CSV

```python
import glob

file_paths = glob.glob("../data/raw/amazon_india_*.csv")
dfs = []

for file in file_paths:
    temp_df = pd.read_csv(file)
    dfs.append(temp_df)

df = pd.concat(dfs, ignore_index=True)
```
